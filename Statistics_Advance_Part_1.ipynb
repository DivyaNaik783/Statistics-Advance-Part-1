Q3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

  Ans:

```python
def bernoulli_pdf(p, x):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    p (float): The probability of success (0 <= p <= 1).
    x (int): The outcome (0 or 1).

    Returns:
    float: The probability of the given outcome.
    """
    if not (0 <= p <= 1):
        raise ValueError("The probability p must be between 0 and 1.")
    if x not in [0, 1]:
        raise ValueError("The outcome x must be either 0 or 1.")

    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7  # Probability of success
x = 1    # Outcome
print(f"Probability of x={x}: {bernoulli_pdf(p, x):.2f}")
```

Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
a = 1  # Minimum value (inclusive)
b = 10  # Maximum value (inclusive)
n = b - a + 1  # Number of outcomes

# Generate the values and their probabilities
values = np.arange(a, b + 1)
pmf = np.ones(n) / n  # Probability Mass Function (PMF) for uniform distribution

# Calculate the Cumulative Distribution Function (CDF)
cdf = np.cumsum(pmf)

# Plot the CDF
plt.figure(figsize=(8, 5))
plt.step(values, cdf, where='post', label="CDF", color="blue", linewidth=2)

# Add markers for clarity
plt.scatter(values, cdf, color="red", zorder=5)

# Add labels and title
plt.title("Cumulative Distribution Function (CDF) of Discrete Uniform Distribution", fontsize=14)
plt.xlabel("Values", fontsize=12)
plt.ylabel("Cumulative Probability", fontsize=12)

# Add grid and legend
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend(fontsize=10)

# Show the plot
plt.show()
```

Q10.Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

  Ans:

```python
import numpy as np
import matplotlib.pyplot as plt

def central_limit_theorem_demo(sample_size, num_samples):
    """
    Demonstrates the Central Limit Theorem using a non-normal distribution.

    Parameters:
    sample_size (int): The size of each sample.
    num_samples (int): The number of samples to generate.
    """
    # Generate data from a non-normal distribution (e.g., exponential distribution)
    original_data = np.random.exponential(scale=2, size=100000)

    # Collect sample means
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(original_data, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))

    # Plot the original distribution
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.hist(original_data, bins=50, density=True, color='lightcoral', edgecolor='black')
    plt.title("Original Distribution (Non-Normal)", fontsize=14)
    plt.xlabel("Value", fontsize=12)
    plt.ylabel("Density", fontsize=12)

    # Plot the sampling distribution of the mean
    plt.subplot(1, 2, 2)
    plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black')
    plt.title(f"Sampling Distribution of the Mean\n(Sample Size = {sample_size})", fontsize=14)
    plt.xlabel("Sample Mean", fontsize=12)
    plt.ylabel("Density", fontsize=12)

    plt.tight_layout()
    plt.show()

# Example usage:
if __name__ == "__main__":
    central_limit_theorem_demo(sample_size=30, num_samples=1000)
```

Q17.Generate random variables and calculate their corresponding probabilities using the binomial distribution.

  Ans:

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def generate_binomial_distribution(n, p, size):
    """
    Generate random variables from a binomial distribution and calculate their probabilities.

    Parameters:
    n (int): Number of trials.
    p (float): Probability of success on a single trial.
    size (int): Number of random variables to generate.

    Returns:
    np.ndarray: Array of random variables from the binomial distribution.
    np.ndarray: Array of probabilities corresponding to the random variables.
    """
    # Generate random variables from a binomial distribution
    random_variables = np.random.binomial(n, p, size)

    # Calculate probabilities using the PMF (Probability Mass Function) of the binomial distribution
    probabilities = binom.pmf(random_variables, n, p)

    return random_variables, probabilities

def plot_binomial_distribution(n, p, size):
    """
    Plot the binomial distribution for a given number of trials, success probability, and sample size.
    """
    # Generate random variables and their corresponding probabilities
    random_variables, probabilities = generate_binomial_distribution(n, p, size)

    # Plot the histogram of the random variables
    plt.figure(figsize=(8, 5))
    plt.hist(random_variables, bins=np.arange(n+2)-0.5, density=True, alpha=0.7, color='skyblue', edgecolor='black', label="Histogram of Random Variables")

    # Plot the probability mass function (PMF)
    x = np.arange(n+1)
    pmf = binom.pmf(x, n, p)
    plt.plot(x, pmf, 'ro-', label="PMF")

    # Add labels, title, and legend
    plt.title(f"Binomial Distribution (n={n}, p={p})", fontsize=14)
    plt.xlabel("Number of Successes", fontsize=12)
    plt.ylabel("Probability", fontsize=12)
    plt.legend()

    # Show grid and plot
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

# Example usage
if __name__ == "__main__":
    n = 10    # Number of trials
    p = 0.5   # Probability of success on each trial
    size = 1000  # Number of random variables to generate

    plot_binomial_distribution(n, p, size)
```

Q18.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

  Ans:


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    data_point (float): The data point for which to calculate the Z-score.
    mean (float): The mean of the dataset.
    std_dev (float): The standard deviation of the dataset.

    Returns:
    float: The Z-score for the data point.
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_with_standard_normal_distribution(z_score):
    """
    Compare the calculated Z-score with the standard normal distribution.

    Parameters:
    z_score (float): The calculated Z-score for a data point.
    """
    # Standard normal distribution
    x = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(x, loc=0, scale=1)

    # Plot the standard normal distribution
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, color='blue', label='Standard Normal Distribution')

    # Highlight the Z-score on the plot
    plt.axvline(z_score, color='red', linestyle='dashed', label=f'Z-score = {z_score:.2f}')
    plt.fill_between(x, pdf, where=(x >= z_score), color='red', alpha=0.1)

    # Add labels, title, and legend
    plt.title(f"Standard Normal Distribution with Z-score = {z_score:.2f}", fontsize=14)
    plt.xlabel("Z-score", fontsize=12)
    plt.ylabel("Probability Density", fontsize=12)
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Data point and dataset statistics (mean and std_dev)
    data_point = 75
    mean = 70
    std_dev = 10

    # Calculate the Z-score
    z_score = calculate_z_score(data_point, mean, std_dev)

    # Compare with the standard normal distribution
    compare_with_standard_normal_distribution(z_score)
```

Q19.Implement hypothesis testing using Z-statistics for a sample dataset

  Ans:

```python
import numpy as np
import scipy.stats as stats

def z_test(sample, population_mean, population_std):
    """
    Perform Z-test for hypothesis testing.

    Parameters:
    sample (array-like): The sample data.
    population_mean (float): The population mean under the null hypothesis.
    population_std (float): The population standard deviation.

    Returns:
    tuple: Z-statistic, p-value, and conclusion based on the hypothesis test.
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Standard error of the mean
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value for a two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

    # Hypothesis decision (significance level = 0.05)
    if p_value < 0.05:
        conclusion = "Reject the null hypothesis (H). The sample mean is significantly different from the population mean."
    else:
        conclusion = "Fail to reject the null hypothesis (H). The sample mean is not significantly different from the population mean."

    return z_statistic, p_value, conclusion

# Example usage
if __name__ == "__main__":
    # Sample data
    sample_data = [70, 72, 68, 74, 71, 73, 69, 75, 70, 72]

    # Population parameters
    population_mean = 70  # Population mean under the null hypothesis
    population_std = 5    # Population standard deviation

    # Perform Z-test
    z_statistic, p_value, conclusion = z_test(sample_data, population_mean, population_std)

    # Output results
    print(f"Z-statistic: {z_statistic:.3f}")
    print(f"P-value: {p_value:.4f}")
    print(conclusion)
```

Q20.Create a confidence interval for a dataset using Python and interpret the result.

  Ans:

```python
import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(sample, confidence_level):
    """
    Calculate the confidence interval for the mean of a sample.

    Parameters:
    sample (array-like): The sample data.
    confidence_level (float): The desired confidence level (e.g., 0.95 for 95% CI).

    Returns:
    tuple: The lower and upper bounds of the confidence interval.
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation with Bessel's correction

    # Degrees of freedom
    df = sample_size - 1

    # t-critical value for the given confidence level
    t_critical = stats.t.ppf((1 + confidence_level) / 2, df)

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Margin of error
    margin_of_error = t_critical * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
if __name__ == "__main__":
    # Sample data
    sample_data = [70, 72, 68, 74, 71, 73, 69, 75, 70, 72]

    # Desired confidence level
    confidence_level = 0.95  # 95% confidence interval

    # Calculate the confidence interval
    lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level)

    # Output the result
    print(f"The {confidence_level*100}% confidence interval for the mean is: ({lower_bound:.2f}, {upper_bound:.2f})")
```

Q22.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

  Ans:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_normal_pdf(mu, sigma):
    """
    Calculate and plot the probability density function (PDF) of a normal distribution.

    Parameters:
    mu (float): The mean of the normal distribution.
    sigma (float): The standard deviation of the normal distribution.
    """
    # Generate a range of x values (from -4*sigma to +4*sigma around the mean)
    x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

    # Calculate the PDF of the normal distribution
    pdf = stats.norm.pdf(x, loc=mu, scale=sigma)

    # Plot the normal distribution
    plt.figure(figsize=(8, 5))
    plt.plot(x, pdf, label=f'Normal Distribution\n$\mu={mu}$, $\sigma={sigma}$', color='blue', lw=2)

    # Highlight the mean with a vertical line
    plt.axvline(mu, color='red', linestyle='dashed', label='Mean')

    # Add labels, title, and legend
    plt.title("Probability Density Function (PDF) of Normal Distribution", fontsize=14)
    plt.xlabel("x", fontsize=12)
    plt.ylabel("Probability Density", fontsize=12)
    plt.legend()

    # Show grid and plot
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Parameters for the normal distribution
    mean = 0    # Mean
    std_dev = 1  # Standard deviation

    # Plot the PDF of the normal distribution
    plot_normal_pdf(mean, std_dev)

```

Q23.Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

  Ans:

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def plot_poisson_cdf(lam, max_x):
    """
    Calculate and plot the cumulative distribution function (CDF) of a Poisson distribution.

    Parameters:
    lam (float): The mean rate (lambda) of the Poisson distribution.
    max_x (int): The maximum value of x for which to calculate the CDF.
    """
    # Generate x values from 0 to max_x
    x = np.arange(0, max_x + 1)

    # Calculate the CDF for each x
    cdf = stats.poisson.cdf(x, mu=lam)

    # Plot the CDF of the Poisson distribution
    plt.figure(figsize=(8, 5))
    plt.step(x, cdf, where="post", label=f'Poisson CDF\n$\lambda={lam}$', color='blue', lw=2)

    # Add labels, title, and legend
    plt.title("Cumulative Distribution Function (CDF) of Poisson Distribution", fontsize=14)
    plt.xlabel("x", fontsize=12)
    plt.ylabel("Cumulative Probability", fontsize=12)
    plt.legend()

    # Show grid and plot
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Parameters for the Poisson distribution
    lam = 3  # Mean rate (lambda)
    max_x = 10  # Maximum x value to plot

    # Plot the Poisson CDF
    plot_poisson_cdf(lam, max_x)
```

Q25.Write a Python program to compare the standard deviations of two datasets and visualize the difference.

  Ans:

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    data1 (array-like): The first dataset.
    data2 (array-like): The second dataset.
    """
    # Calculate the standard deviations of both datasets
    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation (Bessel's correction)
    std_dev2 = np.std(data2, ddof=1)

    # Print the standard deviations
    print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

    # Plot the datasets using histograms
    plt.figure(figsize=(10, 6))

    # Plot histogram for Dataset 1
    plt.hist(data1, bins=20, alpha=0.5, label=f'Dataset 1 (std={std_dev1:.2f})', color='blue')

    # Plot histogram for Dataset 2
    plt.hist(data2, bins=20, alpha=0.5, label=f'Dataset 2 (std={std_dev2:.2f})', color='red')

    # Add labels and title
    plt.title('Comparison of Standard Deviations', fontsize=14)
    plt.xlabel('Value', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    plt.legend()

    # Show grid and plot
    plt.grid(True)
    plt.show()

# Example usage
if __name__ == "__main__":
    # Generate two datasets with different standard deviations
    data1 = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10
    data2 = np.random.normal(loc=50, scale=5, size=1000)   # Mean=50, Std=5

    # Compare the standard deviations and visualize the difference
    compare_standard_deviations(data1, data2)
```

Q26.Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

  Ans:

```python
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a dataset.

    Parameters:
    data (array-like): The dataset.

    Returns:
    tuple: The range and interquartile range (IQR).
    """
    # Calculate the range (max - min)
    data_range = np.max(data) - np.min(data)

    # Calculate the interquartile range (IQR)
    Q1 = np.percentile(data, 25)  # 25th percentile
    Q3 = np.percentile(data, 75)  # 75th percentile
    IQR = Q3 - Q1

    return data_range, IQR

# Example usage
if __name__ == "__main__":
    # Generate a dataset from a normal distribution
    mean = 50  # Mean of the distribution
    std_dev = 10  # Standard deviation of the distribution
    size = 1000  # Number of data points

    # Generate the dataset
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate the range and IQR
    data_range, IQR = calculate_range_and_iqr(data)

    # Print the results
    print(f"Range of the dataset: {data_range:.2f}")
    print(f"Interquartile Range (IQR) of the dataset: {IQR:.2f}")
```

Q27.Implement Z-score normalization on a dataset and visualize its transformation.

  Ans:

```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization (standardization) on a dataset.

    Parameters:
    data (array-like): The dataset to normalize.

    Returns:
    array-like: The normalized dataset (mean=0, std=1).
    """
    # Calculate the mean and standard deviation
    mean = np.mean(data)
    std_dev = np.std(data)

    # Apply the Z-score formula
    normalized_data = (data - mean) / std_dev

    return normalized_data

# Example usage
if __name__ == "__main__":
    # Generate a dataset from a normal distribution
    mean = 50  # Mean of the distribution
    std_dev = 10  # Standard deviation of the distribution
    size = 1000  # Number of data points

    # Generate the dataset
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Normalize the dataset using Z-score
    normalized_data = z_score_normalization(data)

    # Plot the original and normalized data
    plt.figure(figsize=(12, 6))

    # Plot the original data
    plt.subplot(1, 2, 1)
    plt.hist(data, bins=30, color='blue', alpha=0.7)
    plt.title("Original Data (Mean=50, Std=10)")
    plt.xlabel("Value")
    plt.ylabel("Frequency")

    # Plot the normalized data
    plt.subplot(1, 2, 2)
    plt.hist(normalized_data, bins=30, color='green', alpha=0.7)
    plt.title("Normalized Data (Mean=0, Std=1)")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")

    # Show the plots
    plt.tight_layout()
    plt.show()
```

Q27.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

  Ans:


```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    """
    Calculate the skewness and kurtosis of a dataset.

    Parameters:
    data (array-like): The dataset to analyze.

    Returns:
    tuple: The skewness and kurtosis of the dataset.
    """
    # Calculate the skewness and kurtosis
    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)

    return data_skewness, data_kurtosis

# Example usage
if __name__ == "__main__":
    # Generate a dataset from a normal distribution
    mean = 50  # Mean of the distribution
    std_dev = 10  # Standard deviation of the distribution
    size = 1000  # Number of data points

    # Generate the dataset
    data = np.random.normal(loc=mean, scale=std_dev, size=size)

    # Calculate the skewness and kurtosis
    skewness, kurt = calculate_skewness_and_kurtosis(data)

    # Print the results
    print(f"Skewness of the dataset: {skewness:.4f}")
    print(f"Kurtosis of the dataset: {kurt:.4f}")
```
